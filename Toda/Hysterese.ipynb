# Hysterese bei Untersuchung der Dynamik und Bifurkationsdiagrammen

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hysterese(rise,fall,title='Hysteresediagramm'):
    '''Nimmt die Positionen der Bifurkationspunkte und plottet Hysteresediagramm'''
    ypos = np.arange(0,8)
    yticks = np.arange(0,9)
    yticklabels = ['1','2','4','Chaos','5','Chaos','3','6','Chaos']

    ax = plt.subplots()

    plt.step(rise,ypos,label='steigende Spannung',color='C0')
    plt.step(fall,ypos,label='fallende Spannung',color='C1')
    plt.step([0,rise[0]],[0,0],color='C0')
    plt.step([rise[-1],13],[7,8],color='C0')
    plt.step([0,fall[0]],[0,0],color='C1')
    plt.step([fall[-1],13],[7,8],color='C1')

    plt.xlim(0,13)
    plt.ylim(0,8)
    plt.xlabel(r'Amplitudenspannung in $\mathrm{V}_{\mathrm{pp}}$')
    plt.ylabel('Zahl der Zustände')
    plt.yticks(yticks,yticklabels)
    plt.grid()
    plt.legend()
    plt.title(title)
```

Da sollen eventuell noch achtfacher und zwölfacher Zustand vor erstem Chaos sein...  
Nutze die am Oszi gemessene Amplitude...  
Für Untersuchung der Dynamik ist Diagramm Müll...  
Untersuchung der Dynamik: 

```python
rise = np.array([1.88,4.94,5.76,6.72,6.8,8.16,10.9,12.4])
fall = np.array([2.0,5.04,5.84,6.72,6.8,7.6,10.7,11.9])

# genauen Parameter mit in Titel
hysterese(rise,fall)
plt.savefig('figures/Hysteresediagramm_erste_Untersuchung.pdf')
```

```python
def hysterese_2(rise,fall,title='Hysteresediagramm'):
    '''Nimmt die Positionen der Bifurkationspunkte und plottet Hysteresediagramm'''
    ypos = np.arange(0,6)
    yticks = np.arange(0,7)
    yticklabels = ['1','2','4','Chaos','3','6','Chaos']

    ax = plt.subplots()

    plt.step(rise,ypos,label='steigende Spannung',color='C0')
    plt.step(fall,ypos,label='fallende Spannung',color='C1')
    plt.step([0,rise[0]],[0,0],color='C0')
    plt.step([rise[-1],13],[5,6],color='C0')
    plt.step([0,fall[0]],[0,0],color='C1')
    plt.step([fall[-1],13],[5,6],color='C1')

    plt.xlim(0,13)
    plt.ylim(0,6)
    plt.xlabel(r'Amplitudenspannung in $\mathrm{V}_{\mathrm{pp}}$')
    plt.ylabel('Zahl der Zustände')
    plt.yticks(yticks,yticklabels)
    plt.grid()
    plt.legend()
    plt.title(title)
```

Nun Hysteresediagramme aus Bifurkationsdiagrammen:  
Teilweise echte Willkür beim Wählen der Punkte. Ergebnisse rein qualitativ, keine quantitativen Aussagen...  
Fünfacher Zustand und folgendes Chaos nicht zu erkennen...

```python
# Noch Plotfunktion schreiben, die erkennbare Zustände aus Bifurkationsdiagramm plottet...
# Fehler entsprechend Auflösung des Oszis bei 0.2 V

rise1 = np.array([0.9,2.5,2.9,4.0,4.4,6.0])
fall1 = np.array([0.9,2.4,2.9,3.7,5.3,6.0])
rise5 = np.array([0.9,2.5,2.9,4.1,5.5,5.9])
fall5 = np.array([0.8,2.4,2.9,3.7,5.3,5.9])
rise30 = np.array([1.0,2.6,3.5,4.0,5.8,7.0])
fall30 = np.array([0.9,2.2,2.7,3.6,4.0,5.6])
rise100 = np.array([1.5,3.4,4.0,4.4,6.2,7.0])
fall100 = np.array([0.7,2.2,2.8,3.6,4.8,5.8])
rise200 = np.array([1.3,3.2,3.8,4.3,6.3,7.0])
fall200 = np.array([0.7,2.0,3.0,3.4,4.6,6.0])
rise300 = np.array([1.4,3.2,4.0,4.5,6.4,7.5])
fall300 = np.array([0.8,2.0,2.9,3.3,4.6,5.2])

freq = np.array([1,5,30,100,200,300])
r = np.array([rise1,rise5,rise30,rise100,rise200,rise300])
f = np.array([fall1,fall5,fall30,fall100,fall200,fall300])

for i in range(len(r)):
    hysterese_2(r[i],f[i],title='Frequenz $f='+str(freq[i])+'$ Hz')
    plt.savefig('figures/hysterese_'+str(freq[i])+'.pdf')
```
