# Impulsformen
Impulsgenerator:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

pulsheight = np.array([1.,2.,3.,4.,5.,6.])
puls_volt = np.array([24.,78.,116.,150.,184.,220.])
normalize = np.array([1.,2.,3.,4.,5.,6.])
norm_volt = np.array([80.,122.,150.,162.,176.,184.])
```

```python
fig, ax = plt.subplots()

ax.scatter(pulsheight,puls_volt,c='red',label='PULS HEIGHT')
ax.scatter(normalize,norm_volt,c='blue',label='NORMALIZE')

slope, intercept, r, p, se = linregress(pulsheight, puls_volt)
func = lambda x: slope * x + intercept
fit = list(map(func, pulsheight))

ax.plot(pulsheight, fit, c='red')
print('slope: ', slope)
print('intercept: ', intercept)
print('Pearson coefficient: ', r)

ax.set_xlabel('Parameter in willk√ºrlichen Einheiten')
ax.set_ylabel('Spannung in mV')

plt.title('Impulsgenerator')

plt.legend()
plt.tight_layout()
plt.grid()
plt.savefig('plots/Lin_impulsgenerator.pdf')
```
